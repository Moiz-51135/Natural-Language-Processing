# Removing  Html Tags

```python
import re
def remtag(data):
    p=re.compile(r'<.*?>')
    return p.sub('',data)
```

```python
remtag(sample_text)
```

# Sentence Tokenization

```python
dummy_para="In a small coastal town nestled between rolling hills and the shimmering sea, lived a curious young girl named Amelia. Her days were filled with the enchantment of exploring hidden coves, collecting seashells, and listening to the tales of old fishermen. One day, while wandering along the shore, she discovered a mysterious, weathered map washed ashore. The faded lines and cryptic symbols hinted at a long-forgotten treasure buried on a distant island. Fueled by excitement and a sense of adventure, Amelia embarked on a thrilling journey, encountering challenges, forging unexpected friendships, and unlocking the secrets of the map that would forever change the course of her life."
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
sents=sent_tokenize(dummy_para)
sents
```

# Word Tokenization

```python
for sent in sents:
     print(word_tokenize(sent))
```
