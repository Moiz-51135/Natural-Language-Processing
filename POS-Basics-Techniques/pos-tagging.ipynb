```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc_1=nlp(u"Hey! how is life going on?")
```

```python
doc_1.text
```

```python
doc_1[0]
```

# Coarse Grained POS

```python
doc_1[0].pos_
```

# Fine Grained POS

```python
doc_1[0].tag_
```

```python
spacy.explain('UH')
```

```python
spacy.explain('INTJ')
```

```python
for word in doc_1:
    print(word.text,"------>", word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc_2=nlp(u"I came to office by time")
for word in doc_2:
    print(word.text,"------>",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc_3=nlp(u"I will come to office by time")
for word in doc_3:
    print(word.text,"------>",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc_4=nlp(u"i am done with NLP and currently i am gaining my skills in LLMs")
for word in doc_4:
    print(word.text,"----->",word.pos_,word.tag_,spacy.explain(word.tag_))
```
