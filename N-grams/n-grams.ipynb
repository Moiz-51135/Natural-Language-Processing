```python
import numpy as np
import pandas as pd
```

```python
data = {
    'text': [
        "I love this product!",
        "The service was terrible.",
        "The movie was amazing.",
        "I feel neutral about it.",
        "The experience was disappointing.",
        "This is the best thing ever!",
        "I can't stand this.",
    ],
    'output': [
        1,
        0,
        1,
        0,
        0,
        1,
        0
    ]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

# Uni-grams

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1, 1))
vectorizer_lowercase = CountVectorizer(lowercase=True, stop_words='english')
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

# Vocabulary Count

```python
import collections

all_text = ' '.join(df['text'])
words = all_text.split()
vocabulary = collections.Counter(words)

unique_vocab_count = len(vocabulary.keys())
print("Count of Unique Vocabulary:", unique_vocab_count)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
print(bow[4].toarray())
```
