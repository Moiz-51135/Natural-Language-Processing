```python
import fitz  # PyMuPDF for working with PDFs
import re

def extract_information_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)

    # Initialize variables to store information
    date = None
    statement_reference = None
    contract = None
    activity_performer = {}

    # Iterate through pages
    for page_num in range(pdf_document.page_count):
        # Extract text from the current page
        page_text = pdf_document[page_num].get_text()

        # Use regex to find patterns in the text
        date_match = re.search(r'Date (\d{2}/\d{2}/\d{2})', page_text)
        statement_ref_match = re.search(r'Statement Reference (\d+/\d+)', page_text)
        contract_match = re.search(r'Contract\s*([\s\S]+?)Date', page_text)

        # Update variables if matches are found
        if date_match:
            date = date_match.group(1)
        if statement_ref_match:
            statement_reference = statement_ref_match.group(1)
        if contract_match:
            # Extract lines following the "Contract:" pattern
            contract_lines = [line.strip() for line in contract_match.group(1).splitlines() if line.strip()]
            # Join the contract lines into a single string
            contract = ' '.join(contract_lines)

        # Extract Units of Dental Activity per Performer
        performer_activity_match = re.findall(r'(\d{6})\s+([^\d]+)\s+Current Financial Year (\d+\.\d+)', page_text)
        for performer_id, performer_name, activity_value in performer_activity_match:
            activity_performer.setdefault(performer_id, []).append({
                'performer_name': performer_name.strip(),
                'activity_value': float(activity_value)
            })

    # Close the PDF document
    pdf_document.close()

    return date, statement_reference, contract, activity_performer

# Replace 'your_pdf_path.pdf' with the actual path to your PDF file
pdf_path = '/content/Att8820166.pdf'
date, statement_reference, contract, activity_performer = extract_information_from_pdf(pdf_path)

# Print the extracted information
print(f"Date: {date}")
print(f"Statement Reference: {statement_reference}")
print(f"Contract: {contract}")
print("Units of Dental Activity per Performer:")
for performer_id, activities in activity_performer.items():
    print(f"{performer_id} {activities[0]['performer_name']}")
    print(f"Current Financial Year {date} {activities[0]['activity_value']}")
    print(f"Last Financial Year 2022/23 {activities[1]['activity_value']}")
    print(f"Other Financial Years {activities[2]['activity_value']}\n")
```
